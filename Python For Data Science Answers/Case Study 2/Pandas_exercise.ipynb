## İkinci Kısım

```python
import numpy as np
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
df.groupby("time").agg(["sum", "min", "max", "mean"])["total_bill"]
```

```python
df.groupby(["day", "time"]).agg(["sum", "min", "max", "mean"])["total_bill"]
```

```python
df.groupby(["day", "time"]).agg({"total_bill": ["sum", "min", "max", "mean"], "tip": ["sum", "min", "max", "mean"]})
```

```python
df.loc[(df["size"] < 3) & (df["total_bill"] > 10)].mean()
```

```python
df["total_bill_tip_sum"] = df["total_bill"] + df["tip"]
df.head()
```

```python
df.sort_values("total_bill_tip_sum", ascending = False).head(30)
```
